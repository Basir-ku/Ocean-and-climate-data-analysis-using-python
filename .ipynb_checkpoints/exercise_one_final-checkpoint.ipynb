Make a plot showing boxplot of 500mbar zonal wind for each value of latitude. From the plot determine the latitudes with largest variance in zonal wind. Use wind data from May2000-uvt.nc.

Hints: use conversion to pandas trick and keyword arguments x=, y= of boxplot(). Initially the x axis will look ugly since seaborn shows a label for each boxplot. To improve the appearance you can use the following trick, which manually switches most of the labels off:

```python
from scipy import stats as st
from matplotlib import gridspec
import xarray as xr
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read file
fname='May2000-uvt.nc'
ds = xr.open_dataset(fname)
```

Make a plot showing Boxplot of 500mbar zonal wind for each value of latitute

```python
#Converting xarray to pandas.DataFrame
wu=ds.u.sel(level=500.0, method="nearest")
df=wu.to_dataframe()
df1=df.reset_index(level=['latitude','time'])
```

```python
#df1=df.reset_index(level=['latitude','time'])
```

```python
ax=sns.boxplot(x="latitude",y ="u",data=df1, width=0.5)

for ind, label in enumerate(ax.get_xticklabels()):
    if ind % 10 == 0:  # every 10th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)
```

# from the plot is shows largest varince is at 40 degree south
