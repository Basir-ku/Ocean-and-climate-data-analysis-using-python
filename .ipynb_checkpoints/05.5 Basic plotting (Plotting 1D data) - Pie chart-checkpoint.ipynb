```python
import numpy as np
import matplotlib.pyplot as plt

# Pie chart, the slices will be ordered and plotted counter-clockwise:

labels = "Winter", "Spring", "Summer", "Autumn"
sizes = [14, 20, 25, 31]
explode = (0, 0.1, 0, 0) # only "explode" the 2nd slice (i.e. "Spring")

fig=plt.figure()

plt.pie(sizes, explode=explode, labels=labels,autopct='%1.2f', shadow=True, startangle=90)
plt.title('Seasonal rain share at undisclosed location')
plt.show()
```
