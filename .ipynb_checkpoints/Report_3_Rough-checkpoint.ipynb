##### Exercise 3

Madden-Julian oscillation (MJO). MJO is the major fluctuation in tropical weather on weekly to monthly timescales. One of the most important MJO markers is the zonal wind fluctuation at 200mbar level. Select a longitude $\lambda$ corresponding to the tropical region between Indian Ocean and Americas.

• For the chosen longitude compute a time-series $u(t)$ of 200mbar zonal wind,
averaged between 15N and 15S.

• Fit $u(t)$ with a sin function $u(t) = A + B sin(\omega t + C)$. Plot the fitted function and the data

• Compute $R^2$ for the fit.

• Verify if the fitted frequency works at another longitude, by keeping the same $\omega$ and fitting the other three parameters. Compute $R^2$ for the new fit. Plot the fitted function and data as well.

For data use May2000-uvt.nc

```python
#select for 200 millibar pressure level (closest is 250, closest lon to fiji is 177.5)
fiji250 = ds1.u.sel(level=250.0, longitude=178.0, method="nearest")
fiji250
```

```python
#select for between 15N and 15S (needed to then compute the mean)
fiji250lat = fiji250.sel(latitude=slice(15., -15.))
fiji250lat
```

```python
#calculate the mean over entire latitude range
fiji250latmean = fiji250lat.mean(dim='latitude')
fiji250latmean
```

```python
#plotting the time series
plt.figure(3)
fiji250latmean.plot(color='pink')
plt.title('Time Series of Avg Zonal Wind in Fijis Longitude at 250mb between 15N and 15S')
plt.ylabel('u [m/s]')
plt.show()
```

```python
#set up the number of data points (corresponds to the number of time values)
t = np.arange(1, 32)
t
```

```python
#initial guesses
#A is the offset

guess_mean = (np.mean(fiji250latmean)).values
guess_std = (3*np.std(fiji250latmean)/(2**0.5)/(2**0.5)).values
guess_phase = 0 #C
guess_freq = 1 #omega
guess_amp = 1 #B

print(guess_mean, guess_std)
```

```python
#making the guess better, fitting the sine curve

# we'll use this to plot our first estimate. This might already be good enough for you
data_first_guess = guess_std*np.sin(t+guess_phase) + guess_mean

# Define the function to optimize, in this case, we want to minimize the difference
# between the actual data and our "guessed" parameters
optimize_func = lambda x: x[0]*np.sin(x[1]*t+x[2]) + x[3] - fiji250latmean
est_amp, est_freq, est_phase, est_mean = leastsq(optimize_func, [guess_amp, guess_freq, guess_phase, guess_mean])[0]

# recreate the fitted curve using the optimized parameters
data_fit = est_amp*np.sin(est_freq*t+est_phase) + est_mean

# recreate the fitted curve using the optimized parameters
fine_t = np.arange(1,32,0.1)
#data_fit=est_amp*np.sin(est_freq*fine_t+est_phase)+est_mean
data_fit=est_amp*np.sin(est_freq*t+est_phase)+est_mean
```

```python
#creating the plots of the data, first guess, and sine curve fit

plt.plot(t, fiji250latmean, '.') #plots the data as scatter points
plt.plot(t, data_first_guess, label='first guess')
plt.plot(t, data_fit, label='after fitting')
plt.legend()

plt.xlabel('time step [days]')
plt.ylabel('u [m/s]')
plt.title('Fitting the Sine Curve: Zonal Avg Wind at Fiji Longitude, between 15N and 15S')

plt.show()
```

```python
#compute the R squared 
r2 = r2_score(fiji250latmean,data_fit)
r2
```

```python
#select for surface level wind for the dataset we will need

surfzonal = ds1.sel(level=1000.0, method='nearest')
surfzonal
```

```python
#select one point in the tropics (singapore, lat=1, lon=104 ish)
#closest point is lat =0, lon=105
tp = surfzonal.sel(latitude=1.0, method='nearest')
tp
```

```python
#build a linear model
#select temperature values every day for the longitudes around our point (singapore)

#singapore is at 105 longitude
lonm = 102.5
lonp = 107.5

X = tp.t.sel(longitude=lonm, method='nearest').values
Y = tp.t.sel(longitude=lonp, method='nearest').values
```

```python
#combine the two predictors X and Y into one dataset
Xt = np.concatenate((X[:,np.newaxis],Y[:,np.newaxis]), axis=1)
#Xt = xr.concat([X,Y], "new_dim")
Xt
```

```python
#selecting for the values we want as the outcome, the u at the point 
tp1 = tp.sel(longitude=105.0)

#converting it to a Data Array
#tp1 = tp.to_array(dim='u')

tp1
```

```python
#fit the regression line
regr = linear_model.LinearRegression()
regr.fit(Xt,tp1.u)
```

```python
print('Intercept=',regr.intercept_,'Slope=',regr.coef_)
```

```python
x = np.arange(1,32)
x
```

```python
z1 = [] 
z1 = regr.intercept_ + (regr.coef_[0] * x)
z1

z2 = [] 
z2 = regr.intercept_ + (regr.coef_[1] * x)
z1.shape
```

```python
#add the z1 and z2 to the dataset to plot it later
#this way adds  it as coords
#tp1["z1"] = z1
#tp1["z2"] = z2

tp1["z1"]=(['time'],  z1)
tp1["z2"]=(['time'],  z2)

#check if it is there
tp1
```

```python
ax=sns.scatterplot(x=tp1['time'], y=tp1.u)
#line below wasnt working idk why
#ax.plot(x,z1, label='Z-X linear fit')
plt.xlabel('time')
plt.ylabel('u [m/s]')
plt.legend()
plt.title('Z-X linear regression with multiple predictors')
```

```python
#fit the model to the time series data

#make time series
plt.figure(6)
tp1.u.plot(label='wind speed observations')
tp1.z1.plot(label='with predictor X')
tp1.z2.plot(label='with predictor Y')
plt.legend()
plt.title('Singapore Zonal WInd Predictions')
plt.ylabel('u [m/s]')
plt.show()
```

```python
#compute the R squared 
r2_1 = r2_score(tp1.u,z1)
r2_2 = r2_score(tp1.u,z2)
print(r2_1, r2_2)
```
