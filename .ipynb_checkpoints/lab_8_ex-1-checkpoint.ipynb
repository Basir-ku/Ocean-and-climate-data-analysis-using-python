```python
#lab8 -Ex1
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import scipy.fft as ft
```

```python
fname = 'May2000-uvt.nc'
ds = xr.open_dataset(fname)
```

```python
#for the tropics
u = ds.u.sel(level=500,latitude= -23.4372, method="nearest") #surface zonal wind 
v = ds.v.sel(level=500, latitude = -23.4372 , method="nearest")
t = ds.t.sel(level=500, latitude = -23.4372 , method="nearest")
```

```python
u = u.mean(dim="time").values
v = v.mean(dim="time").values
t = t.mean(dim="time").values
```

```python
#calculate power sprectrum for  u
uhat=ft.rfft(u) # compute real fft of u
ck=2*uhat # to obtain expansion wrt to positive
ck[0]=ck[0]/2 # harmonics multiply u(k) for k>0 by 2
fig1 = plt.figure()
plt.plot(ck*np.conj(ck))
plt.xlim([0,10])
plt.xlabel("k")
plt.ylabel("Power for u")
```

```python
#for v
vhat=ft.rfft(v) # compute real fft of u
ck_v=2*vhat # to obtain expansion wrt to positive
ck_v[0]=ck_v[0]/2 # harmonics multiply u(k) for k>0 by 2
fig2 = plt.figure()
plt.plot(ck_v*np.conj(ck_v))
plt.xlim([0,10])
plt.xlabel("k")
plt.ylabel("Power for v")
```

```python
#for t
that=ft.rfft(t) # compute real fft of u
ck_t=2*that # to obtain expansion wrt to positive
ck_t[0]=ck_t[0]/2 # harmonics multiply u(k) for k>0 by 2
fig3=plt.figure() 
plt.plot(ck_t*np.conj(ck_t))
plt.xlim([0,10])
plt.xlabel("k")
plt.ylabel("Power for t")
```

```python
#energy spectrum 
pi=np.pi
L=pi
x=np.linspace(-L,L,144, endpoint=False)
u2=uhat*np.conj(uhat)
v2=vhat*np.conj(vhat)
ensp=u2+v2
fig4=plt.figure() 
# no need to correct ensp[0] as it will not be plotted
ax1=plt.subplot(1,2,1)
ax1.plot(x,u,label="u")
ax1.plot(x,v,label="v")
ax1.set_xlabel("Longitude [rad] ")
ax1.set_ylabel("$f$")
ax1.set_title("Zonal and meridional velocity")
ax1.legend()
ax2=plt.subplot(122) #same as plt.subplot(1,2,2)
k=np.arange(len(uhat))
ax2.loglog(k[1:],u2[1:],label="$\hat u^2$" )
ax2.loglog(k[1:],v2[1:],label="$\hat v^2$" )
ax2.loglog(k[1:],ensp[1:],label="$E_K(k)$" )
ax2.set_xlabel("$k$")
ax2.set_ylabel("Power")
ax2.set_title("Power and energy spectra")
ax2.legend()
plt.subplots_adjust(wspace=0.4)
```
