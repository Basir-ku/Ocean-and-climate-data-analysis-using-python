```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
```

```python
N=100
t=2
pi=np.pi
w=2*pi
X=np.random.normal(size=N)
Y=0.5+0.5*np.sin((w*t)+0.1)*np.random.normal(size=N)

regr = Pipeline([('sine', PolynomialFeatures(degree=3)),
('linear', LinearRegression(fit_intercept=False))])

regr = regr.fit(X[:,np.newaxis], Y)
Yp=regr.predict(X[:,np.newaxis])


x=np.linspace(np.min(X)-0.5,np.max(X)+0.5,1000)
y=regr.predict(x[:,np.newaxis])

ax=sns.scatterplot(x=X,y=Y)
plt.plot(x,y,'r-',label='u(t) fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

print("Regression coefficients:", regr.named_steps['linear'].coef_)
print("Regression intercept:", regr.named_steps['linear'].intercept_ )
print("R^2: ", regr.score(X[:,np.newaxis],Y))
print("MSE:", mean_squared_error(Y,Yp))
```
