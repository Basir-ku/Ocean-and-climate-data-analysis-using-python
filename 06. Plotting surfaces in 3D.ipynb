```python
# Surface plot example: surface z = sin(x2 + y).

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

pi=np.pi
x=np.linspace(0,pi,100)
y=np.linspace(0,2*pi,200)
(X,Y)=np.meshgrid(x,y)
Z=np.sin(X**2+Y)
ax=plt.axes(projection='3d')
surf=ax.plot_surface(X,Y,Z, cmap='coolwarm') # surf will be needed to create colorbar
ax.view_init(69, 121) # changes the viewing angle
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
plt.colorbar(surf,ax=ax) #let’s add colorbar
plt.show()
```

```python
# Wire frame creation pattern
# ax.plot_wireframe(X, Y, Z, rstride=A, cstride=B)
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
pi=np.pi
x=np.linspace(0,pi,100)
y=np.linspace(0,2*pi,200)
(X,Y)=np.meshgrid(x,y)
Z=np.sin(X**2+Y)
ax=plt.axes(projection='3d')
surf=ax.plot_wireframe(X,Y,Z, rstride=5, cstride=10)
# we will need surf to create colorbar
ax.view_init(69, 121) # changes the viewing angle
plt.show()
```
