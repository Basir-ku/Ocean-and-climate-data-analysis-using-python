Overview
For basic plotting I will use matplotlib library. Import it by import matplotlib.pyplot as plt. Types of plots that we will study in this section

- Line plots for 1D data using plt.plot()

- Contouring and color representation for 2D data. plt.pcolormesh() #color representation of 2D-dependent scalar fields (e.g. Temperature), plt.contour() #contour, plt.contourf() # filled contour, plt.quiver() # for winds and currents

- 3d plotting using mplot3d. See, e.g. figure 14

Plotting 2D data

```python
# Let’s plot a temperature on a fantasy planet on Lon x Lat grid.
import numpy as np
import matplotlib.pyplot as plt
import warnings as wr
wr.filterwarnings('ignore')

lon = np.linspace(0, 360, 361)
lat = np.linspace(-91, 90, 181)
X, Y = np.meshgrid(lon, lat)

pi=np.pi # to save on typing
T=270*(np.cos(Y*pi/180)**2)*(np.cos(X*pi/180)**2)+100.0

plt.pcolormesh(X,Y,T, cmap="coolwarm", vmin=100, vmax=350)
plt.colorbar()
plt.contour(X, Y, T, levels=5)
plt.grid()
plt.xlabel("Longitude[degrees]", fontsize=15)
plt.xlabel("Latitude[degrees]", fontsize=15)
plt.title("Temperature at 850 hPa on a fantasy planet")
plt.show()
```
