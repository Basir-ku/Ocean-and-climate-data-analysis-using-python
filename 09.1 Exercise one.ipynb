Run the supplied script kdv-daao.py that solves Korteveg de Vries (KDV) equation.The script produces 3 binary files:

• kdvx.dat contains spatial grid (values of x)
• kdvt.dat contains temporal grid (values of t)
• kdvsol.dat contains the solution ( u(x, t) ).

Read the files into the python variables and produce the Hovmøller diagram showing evolution u(x,t), similar to figure 19.

Compute the phase speed of the observed waves (the speed of the wave crests).

```python
import numpy as np
import matplotlib.pyplot as plt

# Read the files
xfile = open("kdvx.dat", "rb")
x = np.load(xfile)

tfile = open("kdvt.dat", "rb")
t = np.load(tfile)

ufile = open("kdvsol.dat", "rb")
u = np.load(ufile)

X, T = np.meshgrid(x, t) # Return coordinate matrices from coordinate vectors.

# plot the Hovmøller diagram
plt.contourf(X, T, u)
plt.colorbar()
plt.title("Hovmøller diagram")
plt.grid()
plt.xlabel("t")
plt.ylabel("x")
plt.show()
```
