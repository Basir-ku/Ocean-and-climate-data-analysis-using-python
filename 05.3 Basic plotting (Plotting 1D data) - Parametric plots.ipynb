Example: parametric plot (circle of radius 5)
draw a circle x = R cos(φ) , y = R sin(φ) , φ ∈ [0, 2π]

```python
import numpy as np
import matplotlib.pyplot as plt

pi=np.pi
phi=np.linspace(0,2*pi, 100)

R = 5.0 # let’s draw a circle of radius 5
x=R*np.cos(phi)
y=R*np.sin(phi)
plt.plot(x,y,linewidth=2)
plt.show()
```
