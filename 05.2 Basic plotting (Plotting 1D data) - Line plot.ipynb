```python
import numpy as np
import matplotlib.pyplot as plt

pi=np.pi

x=np.linspace(0,360,30)
y=np.sin(pi*x/180.0)

plt.plot(x,y,'b--o',label='$\sin$',linewidth=2.5)
plt.plot(x,np.cos(pi*x/180.0),'r-d',label='$\cos$',linewidth=0.5)
plt.legend()
plt.title("sin and cos graph")
# additional colour
# plt.plot(x, np.cos(pi*(x+60.0)/180), 'm')
# plt.plot(x, np.cos(pi*x/180), color='darkorange')
# plt.plot(x, np.cos(pi*x/180), color='tab:cyan')
# plt.plot(x, np.cos(pi*x/180), color='tomato')

plt.show()

#plt.close('all')
# plt.plot(x,np.cos(pi*(x+30)/180.0),'r',label='custom line',linestyle=(0,(3,5,1,5,1,5)))
#plt.show()
```
