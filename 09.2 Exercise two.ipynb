Plot the surface

x2 + y2 + z2 = 4.

Hint: Plot it as a parametric surface. Switch to spherical coordinates

x = R cos(t) cos(s)
y = R cos(t) sin(s)
z = R sin(t)

and use meshgrid on vectors t and s to create a non-rectangular grid in (X, Y, Z)
coordinates.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

R = 5
pi = np.pi 

t = np.linspace(0, 2*pi,100)
s = np.linspace(0, 2*pi,100)

T, S = np.meshgrid(t, s)

x = R*np.cos(T)*np.cos(S)
y = R*np.cos(T)*np.sin(S)
z = R*np.sin(T)

ax=plt.axes(projection='3d')

surf=ax.plot_surface(x,y,z, cmap='coolwarm') # surf will be needed to create colorbar
#ax.view_init(69, 121) # changes the viewing angle
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Parametric Surface')

#plt.colorbar(surf,ax=ax) #let’s add colorbar
plt.show()
```
