Branching (If )

Branching example below determines whether randomly generated x is positive or negative and prints the outcome along with the value of x

```python
import random as rd

x=rd.uniform(-1,1) #generate a random number uniformly distributed in [-1,1]

if (x>0):
    print('x=%3.1f is positive' %x)
elif (x<0):
    print('x=%3.1f is negative' %x)
else:
    print('x is zero') #probability of this happening is 0.
```

Loops - For loop iterates over a given sequence.

Example: plot graphs of sin(kx) for k=0,..,3

```python
import numpy as np
import matplotlib.pyplot as plt

pi=np.pi
x=np.linspace(0,2*pi,300)

for i in range(4): # note that range(4)=[0,1,2,3]
    plt.plot(x,np.sin(i*x), label='sin(%dx)' %i, linewidth=2 )
plt.legend()
plt.show()
```

While loop - executes a set of statements as long as a condition is true.

Example: print i as long as i is less than 6.

```python
i = 1
while i < 6:
    print(i)
    i += 1
```

```python
# Continue statement stops the current iteration, and continues with the next
# Example of continue use: prints integers from 1 to 5, skipping 3.
i = 0
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)
```

```python
# break statement stops the loop regardless of other conditions.

for i in range(4): # note that range(4)=[0,1,2,3]
    plt.plot(x,np.sin(i*x), label='sin(%dx)' %i, linewidth=2 )
    if i==2:
        break
plt.legend()
```

Functions

In addition to built-in functions, we can define our own functions. The main reason to do so is to reuse the code instead of copying it.

Example of a user-defined function: computes horizontal windspeed from the zonal and meridional velocity

def wspeed(u,v):
    res=np.sqrt(u**2+v**2)
return res

```python
# Example user-defined function use

import numpy as np
import matplotlib.pyplot as plt

lon=np.linspace(0,360,361)
lat=np.linspace(-90,90,181)

X,Y=np.meshgrid(lon,lat)

pi=np.pi # to save on typing

U=-10*(np.sin(2*Y*pi/180))*(np.cos(X*pi/180)**2)
V=10*(np.cos(Y*pi/180)**2)*(np.sin(2*X*pi/180))

def wspeed(u,v):
    res=np.sqrt(u**2+v**2)
    return res


Z=wspeed(U,V)

plt.pcolormesh(X,Y,Z)
plt.colorbar()
plt.xlabel('Longitude[degrees]')
plt.ylabel('Latitude[degrees]')
plt.title('Windspeed at 850 hPa on a fantasy planet')
plt.show()
```

Scope of variables - The variables defined outside of the function are visible inside it. They are global. However, the variables defined/modified within a function are not accessible outside of it. They are local. Thus, if you need the function to change some value, return it.

```python
# Variable scope demonstration

def my_func():
    x = 10
    a = 3.71
    print("x value inside function:",x)
    print("a value inside function:",a)
    
    
x = 20
my_func()
print("x value outside function:",x)
```

Efficient coding and vectorization

Speed As an interpreted language (the commands are executed by python interpreter as you type them), python is necessarily slower than compiled language, such as C or FORTRAN, which is producing the executable code at compilation time. This is not noticable, when working with small data sets, however as the amount of data grows, the time delay involved might become annoyingly obvious.

Among the various ways of speeding up the data processing, as a rule of thumb, the most efficient is vectorization. It involves avoiding the loops and sticking to matrix operations

So instead of

for i in range(10000000):
    c[i]=a[i]+b[i]
    
use
c=a+b

which is also more elegant and does not depend on the size of the arrays involved. All numpy functions are already vectorized. So, to speed up the code, stick to matrix opera-tions and numpy functions.

Convenience - Having vectorized functions is very convenient. Observe, that np.sin(x) returns an array, when the argument x is an array, which makes plotting a breeze.

User-defined functions that use matrix operations and vectorized functions are also vec-torized. For instance, wspeed function that we defined above also returned an array, without any effort on our part. However, this feature can be easily upset, e.g. by using a comparison



```python
# Example of non-vectorized function: the following code produces an error
import numpy as np
import matplotlib.pyplot as plt

pi=np.pi
x=np.linspace(-10,10,1000)


def func(x):
    if (x < 0):
        res=1.0
    else:
        res=np.cos(x)
    return res


y=func(x)

# The function func works only on scalars, but not on arrays.
```

Fake vectorization This inconvenience can be easily remedied by using function np.vectorize, which takes any scalar returning function and extends it to arrays by applying the function component-wise. For instance, in the above example replace the last line by the following:

vfunc=np.vectorize(func)
y=vfunc(x)

This, however, does not produce fully vectorized code as np.vectorize internally uses loops. So the convenience issue is solved, while the speed issue is not

Proper vectorization Use numpy functions. Define func as follows:

```python
def func1(x):
    res=np.where(x<0,1.0,np.cos(x))
    return res


plt.plot(x,func1(x))
plt.show()
```

Timing example. Compare the performance of np.vectorize and proper vectorization

```python
import numpy as np
import time

pi=np.pi
x=np.linspace(-10,10,10000000)

def func(x):
    if (x<0):
        res=1.0
    else:
        res=np.cos(x)
    return res


def func1(x):
    res=np.where(x<0,1.0,np.cos(x)) # proper vectorizations
    return res


vfunc=np.vectorize(func) # Fake vescorization

stime1=time.time()
y=vfunc(x)
etime1=time.time()

y=func1(x)
etime2=time.time()

print("np.vectorize ------- %f seconds ---" % (etime1 - stime1))
print("Proper vectorization %f seconds ---" % (etime2 - etime1))
```
