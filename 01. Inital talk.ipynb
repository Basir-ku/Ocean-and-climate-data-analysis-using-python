```python
# Script generating a plot with grid, title, legend, and axis labels
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi)

plt.plot(x, np.sin(x), label="$\sin(x)")
plt.plot(x, np.cos(x), label="$\sin(x)")

plt.xlabel("X")
plt.ylabel("f(X)")
plt.title("Tigonometric Functions")
plt.legend()
plt.grid()
plt.savefig("sincos.png")
plt.show()
    
```
