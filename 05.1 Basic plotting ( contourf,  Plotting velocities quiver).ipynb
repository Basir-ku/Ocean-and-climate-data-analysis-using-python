```python
# Let’s plot a temperature on a fantasy planet on Lon x Lat grid.
import numpy as np
import matplotlib.pyplot as plt
import warnings as wr
wr.filterwarnings('ignore')

lon = np.linspace(0, 360, 362)
lat = np.linspace(-90, 90, 181)
X, Y = np.meshgrid(lon, lat)

pi=np.pi # to save on typing
T=270*(np.cos(Y*pi/180)**2)*(np.cos(X*pi/180)**2)+100.0

plt.contourf(X, Y, T, levels=5)
plt.colorbar()
plt.grid()
plt.xlabel('Longitude[degrees]', fontsize=15)
plt.xlabel('Latitude[degrees]', fontsize=15)
plt.title('Temperature at 850 hPa on a fantasy planet')
plt.show()

```

```python
# Contours at custom levels with labels

#fig = plt.figure()
tlev=np.array([100,150, 200, 250, 300, 350])
ax=plt.contour(X,Y,T,levels=tlev)
plt.title('Contours at custom levels')
plt.clabel(ax, inline=1, fontsize=10)

# Now let’s put quiver on the plot

U=-10*(np.sin(2*Y*pi/180))*(np.cos(X*pi/180)**2)
V=10*(np.cos(Y*pi/180)**2)*(np.sin(2*X*pi/180))

step = 15

qv=plt.quiver(X[::step,::step],Y[::step,::step],U[::step,::step],
              V[::step,::step],scale=200.0, color='red') # step is needed to make the arrows less dense

q_typ=10.0 # typical windspeed to show in the legend
Qkey1 = plt.quiverkey(qv, 0.99, -0.10 , q_typ, '%.1f m/s' % q_typ, coordinates='axes')
# plot the legend: (0.95, -0.12) sets the position of the legend arrow.
plt.show()
```
