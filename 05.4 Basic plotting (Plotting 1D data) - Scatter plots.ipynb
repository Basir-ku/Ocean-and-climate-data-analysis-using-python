Scatter plots are used to plot data points on horizontal and vertical axis in the attempt to show how much one variable is affected by another.

plt.scatter use pattern ==> plt.scatter(x,y, s=area, label=’legend_text’)

```python
import numpy as np
import matplotlib.pyplot as plt

#Here we define Temperature and windspeed data for two weather stations
s1_temp=np.array([14.6, 15.7, 16, 17.7, 17.8, 14.4, 12.6, 12, 12.1,
16.2, 13.5, 11.4, 9.5, 9.8, 11.7, 14.2, 13.6, 9.5, 8.9, 10.8,
12.8, 11.3, 5.7, -0.4, 1.9, 10.2, 14.8, 15.6, 14.4, 11.9,13])

s1_wind=np.array([3.6, 0.2,0.3,0.3,0.7,1.6,4.6,9,4.8,1.4,1.7,4.8,6.3,
4.4, 0.8,3.1,2,3.4,5.5,5.2,3,4.8,2.7,1.7,3,2.6,0.8,1.2,
1,5.1,4.2])

s2_temp=np.array([17.2,20,19.9,21.3,21.6,19.8,17.3,17,16.3,19.3,19.3,
17, 14.3, 15.3,17.3,17.8,19.4,15.3,13.9,14.9,17.8,16.3,11.9,6.1,8.3,
13,19.5,20.9,19.4,18.1,16.9])

s2_wind=np.array([2.3,1.9,1.4,1.4,1.6,2.8,2.9,3.4,3.1,1.8,2.2,2.9,4,
3.7, 2.3,2.1,3.1,3.1,3.6,3.2,3.1,2.9,3.5,4.3,4.4,1.8,3.3,
2,2.4,2.9,1.8])

plt.figure()
area=s1_wind*10 #make the size of the dots proportional to windspeed
plt.scatter(s1_temp,s1_wind, s=area, label='Station 1' )
plt.scatter(s2_temp,s2_wind, label='Station 2')
plt.xlabel('Temperature, C')
plt.ylabel('Wind speed, m/s')
plt.legend()
plt.show()
```
